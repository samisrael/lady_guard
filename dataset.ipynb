```python
import pandas as pd
import random

# Step 1: Define the geographic bounding box for random coordinates (latitude and longitude)
# You can adjust the latitude and longitude range to suit the specific area you're targeting.
lat_min, lat_max = 13.0000, 13.2000  # Latitude range
lon_min, lon_max = 80.0000, 80.3000  # Longitude range

# Step 2: Define possible values for features
crime_types = ["Theft", "Assault", "Vandalism", "Burglary"]
street_lighting = ["Good", "Fair", "Poor"]
pedestrian_traffic = ["Low", "Medium", "High"]
cctv_presence = ["Yes", "No"]
sidewalk_condition = ["Good", "Fair", "Poor"]

# Step 3: Generate random data for 1,000 rows
data = []
for _ in range(1000):  # Generate 1,000 rows
    latitude = round(random.uniform(lat_min, lat_max), 6)
    longitude = round(random.uniform(lon_min, lon_max), 6)
    
    crime_type = random.choice(crime_types)
    crime_frequency = random.randint(1, 10)  # Random crime frequency per month
    crime_severity = random.randint(1, 10)  # Random crime severity (1 to 10)
    lighting = random.choice(street_lighting)
    traffic = random.choice(pedestrian_traffic)
    cctv = random.choice(cctv_presence)
    police_proximity = random.randint(100, 2000)  # Distance to nearest police station in meters
    transport_proximity = random.randint(50, 1000)  # Distance to nearest public transport in meters
    sidewalk = random.choice(sidewalk_condition)

    # Append generated row to the dataset
    data.append([latitude, longitude, crime_type, crime_frequency, crime_severity, lighting, 
                 traffic, cctv, police_proximity, transport_proximity, sidewalk])

# Step 4: Create a DataFrame
columns = ['Latitude', 'Longitude', 'CrimeType', 'CrimeFrequency', 'CrimeSeverity', 'StreetLighting', 
           'PedestrianTraffic', 'CCTVPresence', 'PoliceProximity', 'PublicTransportProximity', 'SidewalkCondition']

df = pd.DataFrame(data, columns=columns)

# Step 5: Save the DataFrame to a CSV file
df.to_csv('large_routes_features.csv', index=False)

# Step 6: Output the first 10 rows for confirmation
print(df.head(10))
```
